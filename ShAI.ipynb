```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv(r'C:/Users/nemah samrah/Downloads/Salaries.csv')
df.head()
```

```python
df.columns
```

# 1

```python
num_rows, num_cols = df.shape
print("Number of rows:", num_rows)
print("Number of columns:", num_cols)

data_types = df.dtypes
print("Data types for each column:")
print(data_types)

missing_values = df.isna().sum()
print("Missing values in each column:")
print(missing_values)
```

# 2

```python
salary_stats = df['TotalPay'].describe()
print("Basic salary statistics:")
print(salary_stats)

median_salary = df['TotalPay'].median()
print("Median salary:", median_salary)

mode_salary = df['TotalPay'].mode()[0]  
print("Salary situation:", mode_salary)

min_salary = df['TotalPay'].min()
max_salary = df['TotalPay'].max()
print("Minimum salary:", min_salary)
print("Maximum salary:", max_salary)

salary_range = max_salary - min_salary
print("Salary range:", salary_range)

std_dev_salary = df['TotalPay'].std()
print("Standard deviation of salary:", std_dev_salary)
```

# 3

```python
df_cleaned_rows = df.dropna()

df_cleaned_columns = df.dropna(axis=1)

print("Data after cleaning:")
print(df_cleaned_rows)
print(df_cleaned_columns)
```

# 4

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalPay'], bins=20, kde=True, color='skyblue')
plt.title('Salary distribution')
plt.xlabel('Salary')
plt.ylabel('Number of Employees')
plt.show()

plt.figure(figsize=(8, 8))
df['JobTitle'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('pastel'), startangle=90)
plt.title('Percentage of employees in different departments')
plt.ylabel('')
plt.show()
```

# 5

```python
grouped_data = df.groupby('JobTitle').describe()

print("Summary statistics by department:")
print(grouped_data)

average_salary_by_department = df.groupby('JobTitle')['TotalPay'].mean()
print("Average salaries by department:\n")
print(average_salary_by_department)
```

# 6

```python
salary_age_correlation = df['TotalPay'].corr(df['Year'])

print("The value of the correlation of salary and age:", salary_age_correlation)

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Year', y='TotalPay', color='blue', alpha=0.7)
plt.title('Salary and age relationship')
plt.xlabel('year')
plt.ylabel('salary')
plt.grid(True)
plt.show()
```
